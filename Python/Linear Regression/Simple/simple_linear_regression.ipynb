# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv('/Code/Machine Learning/ml-learning/Python/Linear Regression/Simple/Salary_Data.csv')
X = dataset.iloc[:,:-1].values #Takes all rows and except the last column
y = dataset.iloc[:,-1].values #Takes all label names except the last one which is result
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0) #split 8-2 train-test
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() # Predicts real value, classification predicts category
regressor.fit(X_train,y_train)
```

## Predicting the Test set results

```python
y_predict = regressor.predict(X_test)
print(y_predict)
```

```python
print(y_test)
```

## Visualising the Training set results

```python
plt.scatter(X_train,y_train,color = 'red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title('Salary from experience (Training set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```
